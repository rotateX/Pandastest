```python
import pandas as pd 

f = open('task_1_zhaopin_data.csv', encoding='utf8')
df1 = pd.read_csv(f)
print(df1)
f.close()
```

```python
# 6.选取【学历要求】、【工作经验】、【是否全职】3 列
print(df1[['学历要求', '工作经验','是否全职']].head(5))
```

```python
# 7.将【公司名】设置为 index 索引 (阅读参考文献 7)，并将表格重新命名为 df2;
df2 = df1.set_index("公司名")
print(df2.head(5))
```

```python
# 8.在 df2 上删除指定列【其他备注】(阅读参考文献 8)，并命名新表为 df3;
df3 = df2.drop(['其他备注'], axis=1)
print(df3.head(5))
```

```python
# 9.在 df3 上删除指定行数据，“上海蜗兴科技有限公司”所在行 (阅读参考文献 8)，并将数据命名为 df4;
df4 = df3.drop(['上海蜗兴科技有限公司'], axis=0)
print(df4.head(5))
```

```python
# 保存 df4 为 CSV 格式文件，并把被保存的 CSV 格式文件命名为“task_1_zhaopin_data_update.csv (阅读参考 文献 9)”。

# header=0,不保存列名 index=0 不保存行索引;默认都为true

df4.to_csv('task_1_zhaopin_data_update.csv', encoding='gbk')
```
